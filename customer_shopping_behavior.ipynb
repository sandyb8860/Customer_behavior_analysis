## Observation-: Missing value found in Review Rating 

```python
rating_by_cat= df.groupby('Category')['Review Rating'].median().reset_index()
```

```python
median=rating_by_cat['Review Rating'].median()
```

```python
df['Review Rating']= df['Review Rating'].fillna(median) ## imputing the Null values with median
```

## Standardising the column names

```python
df.columns= df.columns.str.replace(' ','_').str.lower().str.replace('_(usd)','')
```

```python
df.head()
```

## Feature engineering 

```python
## creating a new column age_group which will have different age categories

labels= ['Young Adult','Adult','Middle-Aged','Senior']
df['age_group']= pd.qcut(df['age'],q=4,labels=labels)
```

```python
## Creating a new column purchase frequency days which will have frequency in number of days

df.frequency_of_purchases.value_counts()
```

```python
frequency_mapping= {
            'Weekly': 7,
            'Fortnightly': 14,
            'Monthly': 30,
            'Bi-Weekly': 14,
            'Quarterly': 90,
            'Annually': 365,
            'Every 3 Months':90
    }
df['purchase_frequency_days']= df['frequency_of_purchases'].map(frequency_mapping)
```

```python
df[['purchase_frequency_days','frequency_of_purchases']].head(10)
```
